```python
import urllib
import json
import os
import cv2
```

```python
pharm = []
with open('kukident_batch1_extract_base.json', 'r') as file:
    pharm=file.read()


pharm=pharm.replace('}\n{','}|{')

data=pharm.split("|")
```

```python
classes=[]

for i in range(len(data)):
    dict_raw=json.loads(data[i])
    bounding_boxes=dict_raw['bounding_boxes']
    for z in range(len(bounding_boxes)):
        class_name=bounding_boxes[z]['label']
        classes.append(class_name)

classes=list(set(classes))
print (classes)
with open('classes.txt', 'w') as f:
    for item in classes:
        f.write("%s\n" % item)
```

```python
im_n=0
for i in range(len(data)):
    print ('Downloading image :',i)
    dict_raw=json.loads(data[i])
    idd=dict_raw['_id']['$oid']
    bounding_boxes=dict_raw['bounding_boxes']
    width=dict_raw['width']['$numberInt']
    height=dict_raw['height']['$numberInt']
    url=dict_raw['url']
#     urllib.request.urlretrieve(url,filename='images/'+idd+'.png')
    img = cv2.imread('images/'+idd+'.png')
    
    for z in range(len(bounding_boxes)):
        try:
            class_name=bounding_boxes[z]['label']
            x1=int(bounding_boxes[z]['ltc']['x']['$numberInt'])
            y1=int(bounding_boxes[z]['ltc']['y']['$numberInt'])
            x2=int(bounding_boxes[z]['rbc']['x']['$numberInt'])
            y2=int(bounding_boxes[z]['rbc']['y']['$numberInt'])

            if not os.path.exists('train_data/'+class_name+'/'):
                os.makedirs('train_data/'+class_name)

            crop_img = img[y1:y2, x1:x2]
            resized_image = cv2.resize(crop_img, (200, 500))
            cv2.imwrite('train_data/'+class_name+'/'+str(im_n)+'.png',resized_image)
            im_n=im_n+1
        except:
            print('Error because :',x1,y1,x2,y2)
```
